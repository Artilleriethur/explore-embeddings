# Embeddings erkunden

In diesem Notebook erkundet ihr, wie Sprachmodelle die **Bedeutung** von Wörtern verstehen.

Jedes Wort wird als **Zahlenvektor** (eine Liste von Zahlen) dargestellt. Wörter mit ähnlicher Bedeutung bekommen ähnliche Vektoren und liegen im "Vektorraum" nahe beieinander.

Das Faszinierende: Mit diesen Vektoren kann man **rechnen**!

---
**So geht’s:** Führt die Code-Zellen der Reihe nach aus (Play-Button oder `Shift + Enter`).

## 1. Vorbereitung
Wir laden ein vortrainiertes Modell mit 400.000 englischen Wörtern. Jedes Wort ist als 100-dimensionaler Vektor gespeichert.

*(Das Laden dauert beim ersten Mal ca. 20–30 Sekunden.)*

```python
import gensim.downloader as api

print("Lade Wort-Vektoren (ca. 120 MB) ...")
model = api.load("glove-wiki-gigaword-100")
print(f"Fertig! {len(model)} Woerter geladen, jedes als {model.vector_size}-dimensionaler Vektor.")
```

## 2. Ähnliche Wörter finden

Welche Wörter liegen im Vektorraum am nächsten beieinander?

Ändere einfach das Wort in den Anführungszeichen und führe die Zelle erneut aus!

```python
# Aendere das Wort und fuehre die Zelle erneut aus!
wort = "king"

ergebnis = model.most_similar(wort, topn=8)
print(f"Die 8 aehnlichsten Woerter zu '{wort}':\n")
for w, score in ergebnis:
    balken = "#" * int(score * 30)
    print(f"  {w:<15} {score:.3f}  {balken}")
```

### Probiert es aus!
Ersetzt `"king"` durch andere Wörter, z. B.:
- `"dog"`, `"cat"`, `"pizza"`, `"school"`
- `"germany"`, `"football"`, `"music"`

**Frage:** Sind die Ergebnisse sinnvoll? Was fällt euch auf?

## 3. Wortvektor-Arithmetik

Jetzt wird es spannend: Wir können mit Wortvektoren **rechnen**!

Das berühmteste Beispiel:

> **"king" - "man" + "woman" = ???**

Wir nehmen den Vektor von *king*, ziehen die "männliche Komponente" ab und addieren die "weibliche" — was kommt raus?

```python
# king - man + woman = ???
ergebnis = model.most_similar(
    positive=["king", "woman"],
    negative=["man"],
    topn=5
)

print('"king" - "man" + "woman" = ?')
print("=" * 40)
for i, (w, score) in enumerate(ergebnis):
    marker = "  <-- Treffer!" if i == 0 else ""
    print(f"  {i+1}. {w:<15} ({score:.3f}){marker}")
```

### Erklärung

Das Modell hat aus Millionen von Texten gelernt, dass sich *king* und *queen* genauso zueinander verhalten wie *man* und *woman*.

Das funktioniert auch mit anderen Beziehungen!

```python
# Laender und Hauptstaedte
# paris - france + germany = ???
ergebnis = model.most_similar(
    positive=["paris", "germany"],
    negative=["france"],
    topn=3
)
print('"paris" - "france" + "germany" =')
for w, score in ergebnis:
    print(f"  -> {w} ({score:.3f})")
```

```python
# Verbformen
# walking - walk + swim = ???
ergebnis = model.most_similar(
    positive=["walking", "swim"],
    negative=["walk"],
    topn=3
)
print('"walking" - "walk" + "swim" =')
for w, score in ergebnis:
    print(f"  -> {w} ({score:.3f})")
```

## 4. Jetzt seid ihr dran!

Ändert die Wörter in der nächsten Zelle und probiert eigene Analogien aus.

**Tipps für Analogien:**
- Länder und Sprachen: `"spanish" - "spain" + "japan" = ?`
- Tiere und Jungtiere: `"puppy" - "dog" + "cat" = ?`
- Berufe: `"doctor" - "man" + "woman" = ?`

```python
# === Aendere hier die drei Woerter! ===
wort_plus_1 = "puppy"    # <-- dieses Wort nehmen
wort_minus  = "dog"      # <-- diese Eigenschaft abziehen
wort_plus_2 = "cat"      # <-- diese Eigenschaft addieren

# --- Ab hier nichts aendern ---
ergebnis = model.most_similar(
    positive=[wort_plus_1, wort_plus_2],
    negative=[wort_minus],
    topn=5
)
print(f'"{wort_plus_1}" - "{wort_minus}" + "{wort_plus_2}" =')
print("=" * 40)
for i, (w, score) in enumerate(ergebnis):
    print(f"  {i+1}. {w:<15} ({score:.3f})")
```

### Zusatz 1: Welches Wort passt nicht?

Das Modell kann auch erkennen, welches Wort in einer Gruppe nicht dazugehört:

```python
# Welches Wort passt nicht?
woerter = ["breakfast", "lunch", "dinner", "car"]

aussenseiter = model.doesnt_match(woerter)
print(f"Woerter:     {woerter}")
print(f"Passt nicht: {aussenseiter}")
```

### Zusatz 2: Wie ähnlich sind zwei Wörter?

Wir können direkt messen, wie ähnlich sich zwei Wörter sind (Skala von 0 bis 1):

```python
# Vergleiche Wortpaare
paare = [
    ("dog", "cat"),
    ("dog", "car"),
    ("happy", "joyful"),
    ("happy", "sad"),
    ("king", "queen"),
    ("king", "banana"),
]

print(f"{'Wort 1':<10} {'Wort 2':<10} {'Aehnlichkeit':>12}")
print("-" * 35)
for w1, w2 in paare:
    sim = model.similarity(w1, w2)
    balken = "#" * int(sim * 20)
    print(f"{w1:<10} {w2:<10} {sim:>8.3f}  {balken}")
```
